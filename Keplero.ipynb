```python
def power(list,n):
  return[x**n for x in list]
```

```python
T = [365.24,686.9,4331,10750]
a = [149.6,227.9,778.3,1430]
a_3 = power(a,3)
T_2 = power(T,2)
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.plot(a_3, T_2, 'g',label='Kepler', marker="o")
plt.legend(loc='best')
plt.xlabel('a^3',fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
