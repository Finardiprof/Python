```python
import matplotlib.pyplot as plt
import numpy as np

def calcola_accelerazione(massa, forza):
  """
  Calcola l'accelerazione data la massa e la forza.

  Argomenti:
    massa: La massa dell'oggetto (in kg).
    forza: La forza applicata all'oggetto (in N).

  Restituisce:
    L'accelerazione dell'oggetto (in m/s^2).
  """
  accelerazione = forza / massa
  return accelerazione

def crea_grafico(massa, forza):
  """
  Crea il grafico della legge oraria dell'oggetto.

  Argomenti:
    massa: La massa dell'oggetto (in kg).
    forza: La forza applicata all'oggetto (in N).
  """
  # Calcola l'accelerazione
  accelerazione = calcola_accelerazione(massa, forza)

  # Imposta il tempo massimo e il passo temporale
  tempo_max = 10
  dt = 0.1

  # Crea gli array per tempo e posizione
  tempo = np.arange(0, tempo_max+dt, dt)
  posizione = []

  # Calcola la posizione per ogni istante di tempo
  for t in tempo:
    posizione.append(0.5 * accelerazione * t**2)

  # Crea il grafico
  plt.plot(tempo, posizione)
  plt.xlabel("Tempo (s)")
  plt.ylabel("Posizione (m)")
  plt.title("Movimento dell'oggetto")
  plt.grid(True)
  plt.show()
  return tempo, posizione

def salva_dati(massa, forza, tempo, posizione):
  """
  Salva i dati in un file di testo.

  Argomenti:
    massa: La massa dell'oggetto (in kg).
    forza: La forza applicata all'oggetto (in N).
    tempo: L'array del tempo (in s).
    posizione: L'array della posizione (in m).
  """
  # Crea il nome del file
  nome_file = "dati_movimento.txt"

  # Apre il file in modalit√† scrittura
  with open(nome_file, "w") as file:
    # Scrive i dati nel file
    file.write(f"Massa: {massa} kg\n")
    file.write(f"Forza: {forza} N\n")
    file.write("\n")
    file.write("Tempo\tPosizione\n")
    for t, p in zip(tempo, posizione):
      file.write(f"{t:.2f}\t{p:.4f}\n")

# Esempi di utilizzo
massa = 5  # Massa dell'oggetto (in kg)
forza = 10  # Forza applicata all'oggetto (in N)

# Calcola l'accelerazione
accelerazione = calcola_accelerazione(massa, forza)
print(f"Accelerazione: {accelerazione:.2f} m/s^2")

# Crea il grafico
[T, S] = crea_grafico(massa, forza)

# Salva i dati
salva_dati(massa, forza, T, S)
```
