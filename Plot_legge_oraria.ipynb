```python
# Traccia la curva oraria dalla tabella oraria.

import numpy as np
import matplotlib.pyplot as plt

# crea array t=0,2,4,...28,30
t = np.arange(0,32,2)

# Lista con i valori di s
s = [0.0,255.9,622.3,845.4,1199.5,1418.0,1788.1,1950.3,
     2000.0,1875.8,1662.6,1401.5,961.7,665.4,273.2,0.0]

plt.plot(t, s)               # genera il file grafico
plt.xlabel('tempo [min]')    # aggiunge descrizione asse x
plt.ylabel('posizione [m]')  # aggiunge descrizione asse y
plt.grid()                   # crea reticolo nel piano cartesiano
plt.show()                   # visualizza grafico
```
