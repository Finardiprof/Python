## Esempio su liste e grafici con Matplotlib

**Codice:**

```
# Importa la libreria Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Esercizio 1: Creazione e manipolazione di liste

# Crea una lista di numeri da 1 a 10
numeri = list(range(1, 11))

# Stampa la lista
print(numeri)

# Aggiungi il numero 11 alla fine della lista
numeri.append(11)

# Stampa la lista
print(numeri)

# Rimuovi il numero 5 dalla lista
numeri.remove(5)

# Stampa la lista
print(numeri)

# Calcola la media dei numeri nella lista
media = sum(numeri) / len(numeri)

# Stampa la media
print(media)

# Esercizio 2: Creazione di grafici con Matplotlib

# Crea una lista di numeri da 1 a 10
x = np.linspace(1,10,10)

# Altro modo: Crea una lista di numeri da 1 a 10
xx = np.arange(11)
# NOTA: np.arange(start, stop, step)

# Crea una lista di numeri quadrati dei numeri da 1 a 10
y = x**2

# Crea un grafico a linee con Matplotlib
plt.plot(x, y)

# Aggiungi un titolo al grafico
plt.title("Grafico a linee")

# Aggiungi etichette agli assi x e y
plt.xlabel("x")
plt.ylabel("y")

# Mostra il grafico
plt.show()

# Esercizio 3: Visualizzazione di dati con Matplotlib

# Crea una lista di nomi di città
città = ["Verona", "Roma", "Milano", "Napoli", "Firenze"]

# Crea una lista di temperature per le città
temperature = [20, 25, 18, 22, 21]

# Crea un grafico a barre con Matplotlib
plt.bar(città, temperature)

# Aggiungi un titolo al grafico
plt.title("Temperature nelle città")

# Aggiungi etichette all'asse x
plt.xlabel("Città")

# Aggiungi un'etichetta all'asse y
plt.ylabel("Temperatura (°C)")

# Mostra il grafico
plt.show()

# Esercizio 4: Personalizzazione dei grafici

# Cambia il colore del grafico a linee
#plt.plot(x, y, color="red")

# Cambia il tipo di linea del grafico a linee
#plt.plot(x, y, linestyle="dashed")
```
