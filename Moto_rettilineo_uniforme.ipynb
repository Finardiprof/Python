```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 0
v = 5 # m/s
t = np.linspace(0,10,11)
S = S0 + v*t
V = np.ones(11)
V = v*V
```

```python
plt.plot(t, S)               # genera il file grafico
plt.xlabel('tempo [min]')    # aggiunge descrizione asse x
plt.ylabel('posizione [m]')  # aggiunge descrizione asse y
plt.grid()                   # crea reticolo nel piano cartesiano
plt.show()                   # visualizza grafico
```

```python
plt.plot(t,V)               # genera il file grafico
plt.xlabel('tempo [min]')    # aggiunge descrizione asse x
plt.ylabel('Velocit√† [m/s]')  # aggiunge descrizione asse y
plt.grid()                   # crea reticolo nel piano cartesiano
plt.show()
```
